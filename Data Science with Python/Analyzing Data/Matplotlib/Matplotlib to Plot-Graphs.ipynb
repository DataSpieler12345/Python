```python
import pandas as pd
import matplotlib.pyplot  as plt
```

```python
df = pd.read_csv('ADANIPORTS.NS.csv',parse_dates = True)
df.sort_index(inplace=True)
```

```python
df.head()
```

```python
plt.plot(df['Volume']) #lets plot volumne
```

```python
plt.boxplot(df['Volume']) #Boxplot
```

```python
plt.hist(df['Close']) #Histogram
```

```python
plt.xlabel('Volume')
plt.ylabel('High')
plt.title('Scatter Plot')
plt.scatter(x=df['Volume'],y = df['High'])
```
