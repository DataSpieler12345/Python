# Scikit-Learn

* Scikit is a powerful and modern machine learning Python library 

* It provides
	
    * tools to identify, orgnize, and solve real-life problems
    
    * free downloadable datasets
    
    * libraries to learn and make prediction
    
    * Model support for every problem type
    
    * Model persistence
    
    * Open-source community and vendor support
    

* It is used to build a machine learning model that has features like:

	* Classification
    
    * Regression
    
    * Clustering 
    
    
* The scikit-learn library includes various algorithms like:

	* K-means
    
    * Ke-nearest neighbors
    
    * Support vector machine (SVM)
    
    * Decision trees
    
    
**Scikit-Learn**: Problem-Solution Approach

* Scikit-learn, helps data scientists and machine learning engineers solve problems using the problem-solution approach

	* Selecting a model 
    
    * Finding an estimator object
    
    * Training the model 
    
    * Magin predictions
    
    * Tuning the model
    
    * Finding accuracy
    
    
**Scikit-Learn**: Problem-Solution Considerations

* Points to be considered while working with a scikit-learn dataset or loading data to scikit-learn:

	* Create separate objects for features and responses
    
    * Ensure features and responses only have numeric values
    
    * Verify that the features and responses are in the form of a Numpy ndarray
    
    * Ensure the features and respondes have the same shape and size as the array
    
    * Map the features as x and the responses as y
    
    
**Scikit-Learn**: Prerequisites for installation: The libraries that **must** be installed before installing scikit-learn are:

	* Pandas  
    
    * Numpy
    
    * Scipy
    
    * Matplotlib

    	* !pip install scikit-learn
