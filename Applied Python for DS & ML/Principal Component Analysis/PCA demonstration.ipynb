#### Principal Component Analysis

#### works best with scale data
#### call the SandardScaler to scaling data


```python
from sklearn.preprocessing import StandardScaler 
```

```python
#lets create the object and instance 
ss = StandardScaler()
```

```python
#fit and transform the data 
scaled = ss.fit_transform(iris.data)
```

#### Create the PCA Object

```python
from sklearn.decomposition import PCA 
```

```python
#lets create the object and define the number of compoonents
pca = PCA(n_components=2) #usually we start with 2
```

```python
#again we can call fit and then transform 
ar = pca.fit_transform(scaled) #2D array
```

```python
ar
```

```python
#retrieve the first 5 
ar[:5]
```

```python
#lets create a scatterplot
plt.scatter(ar[:,0],ar[:,1]) #on X axis = first component (o) and Y axis = second component (1)
```

```python
#give some colors
plt.scatter(ar[:,0],ar[:,1], c=iris.target)
```

```python
#give scmap winter
plt.scatter(ar[:,0],ar[:,1], c=iris.target,cmap='winter')
#notice 3 diff clusters
```

```python
#the components 
pca.components_
#principal component were calculated for all four features , so there were two components created
#first component = 0.52106591, -0.26934744,  0.5804131 ,  0.56485654
#second component = 0.37741762,  0.92329566,  0.02449161,  0.06694199
#4 items each component / we had four features / columns
```

#### try to visualize 

```python
#lets create one data frame 
df = pd.DataFrame(pca.components_,columns=iris.feature_names)
df
```

```python
#create a heatmap
sns.heatmap(df)
```

```python
#add a color map
sns.heatmap(df, cmap='RdBu') #to see the positive and negative values 
#notice 0 and 1 are the created components 
#down below, are the four created features
```
