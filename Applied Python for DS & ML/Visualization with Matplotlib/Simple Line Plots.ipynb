```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#to show the graph in the same notebook
%matplotlib inline
```

```python
#linspace / create numbers 1-10 , but 100 time
n = np.linspace(1,10,100)
s = np.sin(n)
c = np.cos(n)
```

```python
plt.plot(n)
```

```python
plt.plot(n,s)
```

```python
plt.plot(n,c)
```

```python
#change the color 
plt.plot(n,c,"orange")
```

```python
#x and y labels
plt.plot(n,c,"orange")
plt.xlabel("Numbers")
plt.ylabel("Values")
```

```python
#title 
plt.plot(n,c,"orange")
plt.xlabel("Numbers")
plt.ylabel("Values")
plt.title("This is the title")
```

```python
plt.plot(n,c,"orange",label="Cosine")
plt.plot(n,s,"purple",label="Sine")
plt.legend()
plt.xlabel("Numbers")
plt.ylabel("Values")
plt.title("This is the title")
#plt.show just if there not %matplotlib inline command
```
