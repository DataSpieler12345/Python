```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#linspace / create numbers 1-10 , but 100 time
n = np.linspace(1,10,100)
s = np.sin(n)
c = np.cos(n)
```

```python
#pyplot interface
plt.plot(n,c,"orange",label="Cosine")
plt.plot(n,s,"purple",label="Sine")
plt.legend()
plt.xlabel("Numbers")
plt.ylabel("Values")
plt.title("This is the title")
```
