#### Piechart

```python
print(df['Date'].dtype)
```

```python
print(df['Time'].dtype)
```

```python
# change format of date and time columns
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce')
```

```python
df['Date']
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%Y', errors='coerce')
```

```python
df['weekday'] = pd.to_datetime(df['Date'],utc=True).dt.day_name()
```

```python
#Date and Time with datetime format
df.info()
```

```python
grp = df.groupby("weekday")
grp.count()
```

```python
grp = df.groupby("weekday")
grp["weekday"].count()
```

```python
grp = df.groupby("weekday")
series = grp["weekday"].count()
names = series.keys()
numbers = series.values
```

```python
fig, ax = plt.subplots()
ax.pie(numbers)
```

```python
fig, ax = plt.subplots()
ax.pie(numbers, labels=names)
plt.show() #tis show just the plot
```

```python
fig, ax = plt.subplots()
ax.pie(numbers, labels=names, autopct='%.1f') #1 decimales places
plt.show() #tis show just the plot
```

```python
fig, ax = plt.subplots()
ax.pie(numbers, labels=names, autopct='%.1f%%') #1 decimales places
plt.show() #tis show just the plot
```
