```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("ggplot") #change the style for all the created plots
```

```python
n = np.arange(1,11)
s = n**2
c = n**3
```

```python
fig, ax = plt.subplots()
ax.plot(n,s)
```

```python
fig, ax = plt.subplots()
ax.plot(n,s, "r^:")
ax.plot(n,c, "bo-", ds="steps")
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.plot(n,s, "r^:")
ax.plot(n,c, "bo-", ds="steps",alpha=0.1) #to reduce the distance and see the red line better
ax.grid()
```

```python
fig, ax = plt.subplots()
ax.plot(n,s, "r^:",alpha=0.1)
ax.plot(n,c, "bo-", ds="steps") #to reduce the distance and see the red line better
ax.grid()
```

```python
fig, ax = plt.subplots()
p = ax.plot(n,s, "r^:")
plt.setp(p) #paramters available
```
