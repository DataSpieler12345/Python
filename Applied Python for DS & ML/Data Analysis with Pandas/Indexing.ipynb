#### Multi indexing

```python
#read the xlsx files and delete the species column
df = pd.read_excel("E:\DATA SETS\Iris_mini.xlsx", index_col=0).drop('Species',axis=1)
df
```

```python
team = 'alpha alpha alpha beta beta beta'
sort = 'one two three one two three'
df['team'] = team.split()
df['sort'] = sort.split()
```

```python
df
```

```python
# save it into a new data frame 
new_df = df.loc[:, ['SL', 'SW', 'PL', 'PW', 'team', 'sort']]
new_df
```

#### Nested index

```python
new_df.set_index(['team','sort'], inplace=True)
new_df
```

```python
# index alpha
new_df.loc['alpha']
```

```python
#select the second row from alpha index
new_df.loc['alpha'].loc['two']
```

```python
#select the second row  and the value of the PW column from alpha index
new_df.loc['alpha'].loc['two']['PW']
```

```python
# select from alpha and beta index the second=two row of sort column
new_df.xs('two',level='sort')
```
