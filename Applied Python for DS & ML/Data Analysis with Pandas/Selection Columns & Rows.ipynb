## Selecting Columns

```python
# Crear DataFrame con los datos
data = {'SepalLengthCm': [51, 49, 70, 64, 63, 58],
        'SepalWidthCm': [35, 30, 32, 32, 33, 27],
        'PetalLengthCm': [14, 14, 47, 45, 60, 51],
        'PetalWidthCm': [2, 2, 14, 15, 25, 19],
        'Species': ['Iris-setosa', 'Iris-setosa', 'Iris-versicolor',
                    'Iris-versicolor', 'Iris-virginica', 'Iris-virginica']}
df = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E', 'F'])

# Cambiar los nombres de las columnas
df = df.rename(columns={'SepalLengthCm': 'SL', 'SepalWidthCm': 'SW',
                        'PetalLengthCm': 'PL', 'PetalWidthCm': 'PW'})

# Mostrar el DataFrame con los nuevos nombres de las columnas
print(df)
```

```python
#select multiple columns
cols = ["SL","PL"]
df[cols]
```

```python
# create a new columns
df["SL2"] = df["SL"]
df
```

```python
df["SPL"] = df["SL"] / df["PL"]
df["SPW"] = df["SW"] / df["PW"]
df
```

```python
df.shape
```

```python
# drop the new columns
df.drop("SL2",axis=1,inplace=True) #axis 1 = column
df
```

```python
df.drop(["SPL","SPW"],axis=1,inplace=True)
df
```
