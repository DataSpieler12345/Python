```python
import pandas as pd
import numpy as np
```

```python
# create a dictionary
data = {
    'first':['Andrea',
            'Paola',
            None,
            'Rolando',
            None,
            'Robert',
            None],
    'last':['Shiffman',
           'Channel',
           'Poison',
           'MISSING',
           'Pereira',
           'Bartolome',
           'MISSING'],
    'age':[53,
          23,
          27,
          35,
          np.nan,
          33,
          np.nan],
    'salary':['52144',
             '102556',
             np.nan,
             'MISSING',
             np.nan,
             '24445',
             np.nan],
}
```

```python
# store the data set into a pandas dataframe 
df = pd.DataFrame(data)
df
```

```python
df.shape
```

```python
#drop NaN values
df.dropna()
```

```python
#check the missing values of column: last
df.dropna(axis=1)
```

```python
df
```

```python
# row 6
df.dropna(how='any')
```

```python
#tresh 
df.dropna(thresh=2)
```

```python
# fill out the nan values
df.fillna(value='-NA-')
```

```python
# the DF | MISSING for NaN | save it and modify the entire DF created
df.replace('MISSING', np.nan, inplace=True)
df
```
