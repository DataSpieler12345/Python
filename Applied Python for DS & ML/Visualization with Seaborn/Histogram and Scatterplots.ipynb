```python
import seaborn as sns
# import matplotlib.pyplot as plt
# %matplotlib inline 
```

```python
#load the data / lib dataset
df = sns.load_dataset('iris')
df.head()
```

```python
#distplot
sns.distplot(df['sepal_length'])
```

```python
sns.distplot(df['sepal_length'],bins=50)
```

```python
#distplot
sns.distplot(df['sepal_length'],bins=30,kde=False)
```

```python
#joint plot
sns.jointplot(x=df['sepal_length'],y=df['petal_length'])
```

```python
#joint plot
sns.jointplot(x=df['sepal_length'],y=df['petal_length'], kind='reg') #reg =regression plot
```

```python
#jointplot
sns.jointplot(x='sepal_length',y='petal_length', kind='reg', data=df)
```

```python
#lmplot
sns.lmplot(x='sepal_length', y='petal_length', data=df)
```

```python
#lmplot
sns.lmplot(x='sepal_length', y='petal_length', data=df,hue='species')
```

```python
#pairplot / very quick analysis of all the data
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='species')
```
