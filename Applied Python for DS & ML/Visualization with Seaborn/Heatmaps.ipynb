```python
import seaborn as sns 
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
#check the statistic of the data set ( 25% / 50% / 75%)
df.describe()
```

```python
#lets create box plot / check the outliers fo sepal_width
sns.boxplot(data=df)
```

```python
#sepal_width outliers
sns.boxplot(x='sepal_width',data=df)
```

```python
sns.boxplot(x=df['sepal_width'])
```

```python
# outliers sepal_width / species
sns.boxplot(x='species', y='sepal_width',data=df)
```

```python
#violin plot
sns.violinplot(x='species', y='sepal_width',data=df)
```

```python
#boxplot and violinplot
sns.boxplot(x='species', y='sepal_width',data=df,color='red')
sns.violinplot(x='species', y='sepal_width',data=df)
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('flights')
df.head()
```

```python
#pivot the data
df = df.pivot_table('passengers', index='month',columns='year')
df.head()
```

```python
#create a heatmap
sns.heatmap(df)
```

```python
sns.heatmap(df, cmap='Reds')
```

```python
sns.heatmap(df, cmap='YlGn')
```

```python
sns.heatmap(df, cmap='OrRd')
```

```python
sns.heatmap(df, cmap='Greens')
```

```python
sns.heatmap(df, cmap='BuGn')
```
