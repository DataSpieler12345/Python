```python
#importing Necessary libraries 
import pandas as pd
```

```python
#load the data
dp = pd.read_csv('data/Titanic.csv')
```

```python
#make a copy of the original dataframe 
df = dp.copy()
```

```python
df.head()
```

```python
# Delete the  "Unnamed" column
df = df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

```python
#df.describe
df.describe()
```

```python
# Mapear valores numéricos a la columna "Sex"
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
# Codificar la columna "Embarked" utilizando one-hot encoding
df = pd.get_dummies(df, columns=['Embarked'])
```

```python
# Corregir el valor "Did not Survived" a "Did not Survive"
df['Survival_Status'] = df['Survival_Status'].replace('Did not Survived', 'Did not Survive')
```

```python
df.head()
```

```python
# Cambiar "Survived" a 1 y "Did not Survived" a 0 en la columna "Survival_Status"
df['Survival_Status'] = df['Survival_Status'].map({'Did not Survived': 0, 'Survived': 1})
```

```python
#Let us see if there is any co-linearity or correlation amongst the variables
df.corr()
```
