```python
import numpy as np
import matplotlib.pyplot as plt

menMeans = [20, 35, 30, 35, 27]
womenMeans = [25, 12, 34, 20, 25]
group = np.arange(5)

plt.bar(group, menMeans, width=0.35, color='r')
plt.bar(group, womenMeans, width=0.35, bottom=menMeans, color='b')

plt.xlabel('Groups')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(group, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend(labels=['Men', 'Women'])
plt.show()
```
