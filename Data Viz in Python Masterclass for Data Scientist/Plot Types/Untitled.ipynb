```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
y = np.array([35,25,25,15])

plt.pie(y)
plt.show()
```

```python
y = np.array([35,25,25,15])
lb = ['Apple', 'Banana', 'Cherry', 'Pear']

plt.pie(y, labels=lb)
plt.show()
```

```python
y = np.array([35,25,25,15])
lb = ['Apple', 'Banana', 'Cherry', 'Pear']

plt.pie(y, labels=lb, startangle=90)
plt.show()
```

```python
y = np.array([35,25,25,15])
lb = ['Apple', 'Banana', 'Cherry', 'Pear']

plt.pie(y, labels=lb, startangle=180)
plt.show()
```

```python
y = np.array([35,25,25,15])
lb = ['Apple', 'Banana', 'Cherry', 'Pear']
exp = [0.2,0,0,0]


plt.pie(y, labels=lb, startangle=180, explode=exp)
plt.show()
```

```python
y = np.array([35,25,25,15])
lb = ['Apple', 'Banana', 'Cherry', 'Pear']
exp = [0.2,0,0,0]


plt.pie(y, labels=lb, startangle=180, explode=exp, shadow=True)
plt.show()
```

```python
y = np.array([35,25,25,15])
lb = ['Apple', 'Banana', 'Cherry', 'Pear']
exp = [0.2,0,0,0]


plt.pie(y, labels=lb)
plt.legend(title='Four Fruits', loc='upper left')
plt.show()
```

```python
y = np.array([35,25,25,15])
lb = ['Apple', 'Banana', 'Cherry', 'Pear']
exp = [0.2,0,0,0]


plt.pie(y)
plt.legend(labels=lb, title='Four Fruits', loc='upper left')
plt.show()
```

```python
langs = np.array(['C', 'C++', 'Java', 'Python', 'PHP'])
students = np.array([23,17,35,29,12])

plt.pie(students, labels=langs, autopct='%1.2f%%')
plt.show()
```
