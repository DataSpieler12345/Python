```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)

ax.plot3D(x,y,z, 'gray')
plt.title('3D line plot')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
c = x + y

ax.plot3D(x,y,z)
plt.title('3D line plot')
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0,1,100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
c = x + y

ax.scatter(x,y,z, c=c)
plt.title('3D line plot')
plt.show()
```

```python
x = np.outer(np.linspace(-2,2,30), np.ones(30))
y = x.copy().T
z = np.cos(x ** 2 + y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x,y,z, cmap='viridis', edgecolor='none')
plt.title('3D line plot')
plt.show()
```
