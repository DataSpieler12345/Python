```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([3,8,1,10])

plt.plot(y, linestyle='-')
plt.show()
```

```python
y = np.array([3,8,1,10])

plt.plot(y, linestyle='-.', color='g')
plt.show()
```

```python
y = np.array([3,8,1,10])

plt.plot(y, linestyle='-.', color='#4CAF50')
plt.show()
```

```python
y = np.array([3,8,1,10])

plt.plot(y, linestyle='-.', color='#4CAF50', linewidth='8')
plt.show()
```

```python
y = np.array([3,8,1,10])
y1 = np.array([6,2,7,11])

plt.plot(y, linestyle='-.', color='g', linewidth='3')
plt.show()
```

```python
y = np.array([3,8,1,10])
y1 = np.array([6,2,7,11])

plt.plot(y)
plt.plot(y1)
plt.show()
```

```python
y = np.array([3,8,1,10])
y1 = np.array([6,2,7,11])

plt.plot(y, linestyle='-.', color='g', linewidth='3')
plt.plot(y1,linestyle='-.', color='r', linewidth='2')
plt.show()
```
