```python
import matplotlib.pyplot as plt
```

```python
y = [1,4,9,16,25,36,49,64]
x1 = [1,16,30,42,55,68,77,88]
x2 = [1,6,12,18,28,40,52,65]

plt.plot(x1, y, 'ys-')
plt.plot(x2, y, 'go--')
plt.legend(labels=('tv', 'Smartphone'), loc='lower right')
plt.grid(True)
plt.title('Ad effects on sales')
plt.xlabel('medium')
plt.ylabel('sales')
plt.show()
```
