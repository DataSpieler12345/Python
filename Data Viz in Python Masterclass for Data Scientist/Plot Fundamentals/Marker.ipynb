```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.array([3,8,1,10])

plt.plot(y, marker='o')
plt.show()
```

```python
y = np.array([3,8,1,10])

plt.plot(y, marker='D')
plt.show()
```

```python
y = np.array([3,8,1,10])

plt.plot(y, marker='D', ms =5)
plt.show()
```

```python
y = np.array([3,8,1,10])

plt.plot(y, marker='D', ms =5, mec ='r')
plt.show()
```

```python
y = np.array([3,8,1,10])

plt.plot(y, marker='D', ms =5, mec ='r', mfc = 'g')
plt.show()
```

```python
y = np.array([3,8,1,10])

plt.plot(y, marker='o', ms =20, mec ='g', mfc = 'g')
plt.show()
```

##### marker | line | color 

```python
y = np.array([3,8,1,10])

plt.plot(y, 'o:r')
plt.show()
```

```python
n = 256

x = np.linspace(-np.pi, np.pi, n, endpoint=True)
y = np.sin(2*x)

plt.plot(x, y, color='blue', alpha=1.00)
plt.fill_between(x, y, 1, color='blue', alpha=.1)
plt.show()
```

```python
n = 256

x = np.linspace(-np.pi, np.pi, n, endpoint=True)
y = np.sin(2*x)

plt.plot(x, y, color='blue', alpha=1.00)
plt.fill_between(x, y, 0, color='blue', alpha=.1)
plt.show()
```
