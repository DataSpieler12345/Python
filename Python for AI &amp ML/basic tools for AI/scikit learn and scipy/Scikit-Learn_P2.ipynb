# Building Your First Model Using Scikit-Learn in Python

Now that you have a good understanding of the shape and distribution of your data, let’s take a look at how building a model works in Scikit-Learn! 

Here are 4 main steps to building a model:

1. Load and split the Data into Training and Testing Data 

2. Build the model 
3. Predict the test data
4. Evaluate your model 

### 1. Load and split the Data into Training and Testing Data 

It’s convention to load the features and the targets into separate variables, X and y respectively. Remember, these values are stored in the .data and .target keys of the Bunch object. 

We can then use the train_test_split() function to split our data into training and testing data.

```python
# Loading and splitting your data
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import metrics

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = load_iris()
X = pd.DataFrame(iris.data)
y = pd.DataFrame(iris.target)

# Take the test size to be 30% and to stratify the data using the targets. 
# The random_state=100 allows you to be able to reproduce this analysis.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)
```

```python
# Let’s take a look at the sizes of them:
print("X_train's shape is ", X_train.shape)
print("X_test's shape is ", X_test.shape)
print("y_train's shape is ", y_train.shape)
print("y_test's shape is ", y_test.shape)
```

The resulting DataFrames are split into 70% and 30% chunks of the former DataFrames. The features DataFrames continue to have four columns, while the target DataFrames only have one.

### 2. Build the model (Decision tree)
Decide on the type of model you’ll use, such as regression or classifier. In this case, you’ll be building a classifier as your resulting labels will be discrete.

Let’s build a model that can be very easily and visually understood. We’ll use a Decision Tree Classifier to model our algorithm.

```python
# Creating your first model

# set the depth of the tree to be 3
# set the random_state to be 100 to reproduce your results. 
model = DecisionTreeClassifier(max_depth=3, random_state=100)
print(type(model))
```

Notice that by now you’ve created the shell of a model, but you have yet modeled the data.

```python
# Fitting your data to a model
model.fit(X_train, y_train)
```

```python
# Visualizing your first decision tree
plt.figure(figsize=(15,6))  # set plot size (denoted in inches)
plot_tree(
    model, 
    feature_names=iris.feature_names,  # input: X values 
    class_names=iris.target_names,     # output y values 
    filled=True,   # filled=True simply instructs Scikit-Learn to color our decision tree plot 
    fontsize=10
)
plt.title('Your First Decision Tree Classifier', size=18, color='grey')
plt.show()
```

Each box in the image above represents a decision point (unless it’s the final box, in which case it represents a decision itself). If the values in the box are match our data point, then the arrow traverses to the left. If the decision does not match the criteria, it moves to the right.

In the case above, only the petal width and length are considered. This is because the depth of our tree is a limiting factor that we set. You can actually check this by using the feature_importances_ attribute of the model. 

```python
# Seeing how important each feature is to the model
print(model.feature_importances_)

# Returns:
# [0.         0.         0.56389932 0.43610068]
```

```python
print(iris.feature_names)
```

This breaks down, out of 100%, how important each feature is to the model. Because both sepal width and length have a value of 0%, they are not considered at all. This verifies what we saw in the tree plot above.

### 3. Predict the test data

Use the model you created above to make predictions on the testing data

```python
# Making predictions with your model
prediction = model.predict(X_test)
print(prediction)
```

### 4. Evaluate your model 

since you have a testing dataset with its true labels, you’re able to calculate a metrics indicating how accurate your model is. This can be done by using the accuracy_score() function of the metrics module, as shown below:

```python
# Evaluating the accuracy of your model
accuracy = metrics.accuracy_score(y_true=y_test, y_pred=prediction)
print(accuracy)
```

From the code above, you can see that the accuracy of the model is 97.8%! This means that the model should be able to predict the species of a flower 97.8% of the time given these values.
