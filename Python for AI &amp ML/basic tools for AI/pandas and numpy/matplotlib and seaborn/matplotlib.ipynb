## What is matplotlib? 
- open-source python library for 2D plotting
- API inspired by MATLAB
- some of the available diagram types: plots, histograms, bar charts, scatterplots ...

## What is pyplot?

matplotlib.pyplot: is a collection of functions that make matplotlib work like MATLAB
each pyplot function makes some change to a figure: e.g. create a figure, create a plotting area in a figure, decorate the plot with labels, etc.

reference: https://matplotlib.org/2.0.2/api/pyplot_api.html

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## How to plot lines?


```python
# define data as lists
x1 = [2,4,6,8]
y1 = [5.5,7,2,4]
x2 = [4,6,8,10,12,14]
y2 = [-12,-14,-7,-12,-3.3,-1]
```

```python
#let's draw the first line
# if no x-value is given an incrementing index, beginning from zero, will be used
plt.plot(y1)
plt.show()
```

```python
# draw x, y plot
plt.plot(x1,y1)
plt.show()
```

```python
#draw two lines in one plot
plt.plot(x1,y1,x2,y2)
plt.show()
```

## Change line/marker style

```python
#-g : line in green 
plt.plot(y1,'-g')

#ro- : red circle with line
plt.plot(x1,y1,'ro-')

#g* : star symbols in green
plt.plot(x1,y1,'ro-',x2,y2,'g*')
plt.show()
```

## Define the range of the axes and add legend 

```python
plt.plot(x2, y2, color='green', linestyle='dashed', marker='o',
     markerfacecolor='blue', markersize=12,label='line1',linewidth=3)
plt.axis([2, 16, -16, 2]) # [xmin, xmax, ymin, ymax]
plt.legend()
plt.show()
```

## Add axes title, axis label and line legend

```python
line1 = plt.plot(x1, y1)
line2 = plt.plot(x2, y2)
plt.setp(line1, color='r', linewidth=2.0, label='mouse')
plt.setp(line2, color='b', linewidth=2.0, label='pig')
plt.title('test results', color='g',fontsize=18)
plt.xlabel('days',fontsize=16,style='italic')
plt.ylabel('value',fontsize=16,style='italic', rotation='horizontal')
plt.legend() # to show legend
plt.show()
```

```python
# change the shape of the figure 
plt.figure(figsize=(8,2)) # (row, column)

line1 = plt.plot(x1, y1)
line2 = plt.plot(x2, y2)
plt.setp(line1, color='r', linewidth=2.0, label='mouse')
plt.setp(line2, color='b', linewidth=2.0, label='pig')
plt.title('test results', color='g',fontsize=18)
plt.xlabel('days',fontsize=16,style='italic')
plt.ylabel('value',fontsize=16,style='italic', rotation='horizontal')
plt.legend() # to show legend
plt.show()
```

## Plot functions and subplots 

```python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
x
```

```python
y1 = np.cos(x)
```

```python
y1
```

```python
plt.plot(x,y1, color='r', linewidth=2.0)
plt.ylabel(r'$cos(x)$', fontsize=16)
plt.show()
```

```python
y2 = np.sin(x)
```

```python
plt.plot(x,y2, color='b', linewidth=2.0)
plt.ylabel(r'$sin(x)$', fontsize=16)
plt.show()
```

```python
# make subplots vertically 
plt.subplot(2, 1, 1)
plt.plot(x, y1, 'yo-')
plt.ylabel(r'$cos(x)$', fontsize=16)
plt.title('Two subplots')

plt.subplot(2, 1, 2)
plt.plot(x, y2, 'r.-')
plt.xlabel('radiant')
plt.ylabel(r'$sin(x)$', fontsize=16)

```

```python
# make subplots horizontally 
plt.subplot(1, 2, 1)
plt.plot(x, y1, 'yo-')
plt.ylabel(r'$cos(x)$', fontsize=14)
plt.xlabel('radiant')

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'r.-')
plt.xlabel('radiant')
plt.ylabel(r'$sin(x)$', fontsize=14)
```

```python
# Improve subplot size/spacing with many subplots in matplotlib
plt.subplot(1, 2, 1)
plt.plot(x, y1, 'yo-')
plt.ylabel(r'$cos(x)$', fontsize=16)
plt.xlabel('radiant')

plt.subplot(1, 2, 2)
plt.plot(x, y2, 'r.-')
plt.xlabel('radiant')
plt.ylabel(r'$sin(x)$', fontsize=16)

plt.tight_layout() # make more space 
```
