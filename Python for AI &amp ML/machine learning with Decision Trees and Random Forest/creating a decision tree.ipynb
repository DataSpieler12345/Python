#### Iris Data set

#### Import libraries and iris dataset


```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree

import warnings
warnings.filterwarnings("ignore")
```

#### The Iris Dataset

```python
iris = load_iris()
X, y = iris.data, iris.target
```

```python
X
```

```python
y
```

```python
np.shape(X)
```

#### Creating the Decision Tree

```python
clf = DecisionTreeClassifier()
```

```python
# fitting the model
clf = clf.fit(X, y)
```

```python
# prediction
clf.predict([[6.1, 2.7, 3.9, 1.2]]) # iris versicolor
```

#### Visualizing the tree

```python
plot_tree(clf)
```

#### We can plot the same graph, but through the Matplotlib Library and we can control the size of our figure

```python
plt.figure(figsize=(15,12))
plot_tree(clf, filled = True)
plt.show()

# x[2] petal width <= 2.45
# gini = 0.667 its a metric of the decision trees
# samples = 150, there are 150 data points at this node
# value = [50,50,50], is the data points in each class
```

```python
plt.figure(figsize=(15,12))
plot_tree(clf, filled = True,
         feature_names = ["Sepal Length","Sepal Width","Petal Length","Petal Widht"],
         class_names = ["Setosa","Versicolour","Virginica"])
plt.show()
```
