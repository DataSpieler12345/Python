# Dropping and editing values

```python
#create a DataFrame
list_of_tuples = [('sfs',2012), ('phx', np.nan), ('phx', 2005), ('chi', 2009)]
df = pd.DataFrame(list_of_tuples, columns=['prioridad', 'edad'])
df['country'] = 'USA'
df['continent'] = 'North America'
df
```

```python
#remove a column from a df with the .drop() by referencing its label and axis
#axis 0 = rows
# axis 1 = columns
df2 = df.drop('country', axis =1, inplace=False)
df2
```

```python
#remove a column frim a dataframe with the del() function
df4 = pd.DataFrame(df2)
del(df4['continent'])
df4
```

```python
#use the len() function to check the row count of a DataFrame
len(df4)
```

```python
#use .count() method to check the row count, but this excludes Nans
df4.count()
```
