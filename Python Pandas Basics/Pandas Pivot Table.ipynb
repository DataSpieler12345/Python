```python
import pandas as pd
import numpy as np
```

```python
#pivot table
```

```python
df = pd.read_excel('sales_funnel.xlsx')
print('Data Loaded')
df
```

```python
df.head() #first 5 rows of the dataset
```

```python
#pivot table
pd.pivot_table(df,index=['Name'])
```

```python
#pass multiple indexises
pd.pivot_table(df,index=['Name','Rep','Manager']) #3 columns index / multiple indexises
```

```python
pd.pivot_table(df,index=['Manager','Rep']) #2 columns indexises
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Quantity'])
```

```python
#aggfunction
pd.pivot_table(df,index=['Manager','Rep'],values=['Quantity'],aggfunc=[np.mean,len])
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Quantity'],aggfunc=np.sum)
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Quantity'],columns=['Product'],aggfunc=[np.sum])
```

```python
df['Product'].unique()
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Quantity','Account'],columns=['Product'],aggfunc=[np.sum],fill_value = 0)
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Quantity'],columns=['Product'],aggfunc=[np.sum],fill_value = 0)
```

```python
pd.pivot_table(df,index=['Manager','Rep'],values=['Quantity','Product'],aggfunc=[np.sum],fill_value = 0)
```
