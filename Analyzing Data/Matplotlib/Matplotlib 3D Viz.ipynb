```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ADANIPORTS.NS.csv',parse_dates = True)
```

```python
df.head()
```

```python
df['H-L'] = df.High - df.Low
df['100MA'] = df['Close'].rolling(100).mean()
```

```python
# Plotting 3D Graph
ax = plt.axes(projection='3d')
ax.scatter(df.index, df['H-L'], df['100MA'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('100MA')
plt.show()
```

```python
# Plotting 3D Graph
ax = plt.axes(projection='3d')
ax.scatter(df.index, df['H-L'], df['Volume'])
ax.set_xlabel('Index')
ax.set_ylabel('H-L')
ax.set_zlabel('Volume')
plt.show()
```
