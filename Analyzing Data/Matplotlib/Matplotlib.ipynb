**Advantages of using matplotlib to visualize data**

* It is a multi-platform data visualization tool; therefore, its fast and efficient.

* It can work well with many operating and graphics systems.

* It has high-quality graphics and plots to print and view for various range of graphs ( such as Histogram, pychart, barchart, scattterplot and heatmaps

* With Jupyter notebook integration, the developers are free to spend their time implementing features. 

* It has large community support and cross platform suppor as it is an open-source tool.

* It has full control over graphs or plot styles.


**Plot**: is a graphical representation of data which shows relationship between two variables or the distribution of data.

**Steps to Create Plot-Example**

```python
#STEP1 = Importe the requeride libraries
#import numpy for generating random numbers
import numpy as np
#import matplotlib library
import matplotlib.pyplot as plt 
from matplotlib import style
%matplotlib inline
```

```python
#STEP2 = Define or import the requeride dataset
#generate random numbers (total 10)
randomNumber = np.random.rand(10)
```

```python
#view them
print(randomNumber)
```

```python
#STEP3 = Set the plot parameters
#select the style of the plot
style.use('ggplot')
#plot the random number
plt.plot(randomNumber, 'g',label='lineone',linewidth=2)
#x axis is number of random numbers (index)
plt.xlabel('range')
#y axis is actual random number
plt.ylabel('Numbers')
#title of the plot
plt.title('First Plot')

plt.legend()
plt.show()
```
